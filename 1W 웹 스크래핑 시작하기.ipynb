## 웹 스크래핑을 소개합니다.
> 이번에는 웹 스크래핑이 무엇인지, 어떤 것을 만들 수 있는지 배워봅시다.

* 웹 스크래핑(Web scraping)은 인터넷에 있는 웹 사이트에서 내가 원하는 정보를 자동으로 모아주거나, 저장해주는 프로그램을 말합니다. 
    * Web은 World Wide Web의 줄임말이며, URL 가장 앞에 붙는 www가 World Wide Web에서 유래했습니다.
    * 인터넷은 컴퓨터 네트워크 통신망을 의미합니다.
    * 웹은 인터넷 상에서 동작하는 서비스를 의미합니다.
* 웹 스크래핑 프로그램을 만들면 데이터를 수집하고, 분석하는데 매우 유용합니다.
    * 예를 들어 상품 가격을 비교하는 웹 사이트에서 상품 가격 정보를 스크래핑하여 비교 정보를 제공하거나, 실시간으로 SNS의 댓글을 수집할 수 있습니다.

### 웹 스크래핑 VS 웹 크롤링
* 웹 스크래핑과 비슷한 개념으로 웹 크롤링(Web crawling)이 있습니다.
    * 웹 스크래핑은 인터넷 상에서 내가 필요한 특정 데이터를 수집하고 분석하는 것에 초점을 맞춘 행위를 말합니다.
    * 웹 크롤링은 인터넷 상에 있는 모든 웹 페이지를 탐색하고 필요한 정보를 수집하는 것에 초점을 맞춘 행위를 말합니다.
    * 웹 크롤링은 웹 크롤러(Web Crawler)라는 일종의 봇(bot)을 이용하여 웹 페이지 안에 있는 모든 페이지를 탐색하며 정보를 수집합니다. (선탐색 후정보추출)
    * 반면, 웹 스크래핑은 내가 데이터를 추출하고 싶은 특정 사이트가 있고, 그 사이트에서 특정 데이터를 수집합니다. (선정보결정 후탐색)

<br>

* 엄밀히 따지자면 둘은 분리되어야 하는 개념이지만, 편의상 둘을 구분하지는 않겠습니다.
    * 두 행위에 사용되는 Python 스킬도 차이가 거의 없습니다.

### 웹 스크래핑과 윤리
* 웹 스크래핑은 특정 웹 사이트에 접속하고, 내가 원하는 데이터를 내려받는 행위가 전부입니다. 따라서 기본적으로 불법은 아닙니다.
    * 하지만 불법으로 간주되는 경우가 정말 많으니, 이 기술은 매우 조심해서 사용해야 합니다. (특히 회사)
    * 뉴스 기사 : [여기어때가 야놀자 숙박정보 대량 빼갔다...2심도 ‘크롤링’ 불법 판정](http://www.newsroad.co.kr/news/articleView.html?idxno=18503)
    * 해당 사례를 간단히 요약하자면, 여기어때가 야놀자 데이터를 무단으로 수집해서 자기 회사의 데이터로 복제 후 이용하고 불법으로 판결이 났습니다.

<br>

* 웹 스크래핑은 수집한 데이터를 상업적으로 이용하는 경우 혹은, 내 프로그램이 상대 서버에 부하를 주는 경우 불법 행위로 간주됩니다.
    * 상업적 목적이 아니라고 하여도 합법이라는 뜻은 아닙니다.

<br>

* 기본적으로 모든 웹 사이트들은 이런 웹 스크래핑 행위에 대한 규제 여부를 명시해놓습니다.
    * 각 사이트별로 스크래핑, 크롤링 행위에 대한 여부는 robots.txt라는 파일에 작성해둡니다.
    * 이와 관련한 내용은 추후에 웹 스크래핑을 실습하며 확인해보겠습니다.

### 2. VS Code 설치하기
* VS Code(Visual Studio Code, 이하 vscode)는 코딩할 때 다양한 편리 기능들을 제공하는 프로그램입니다.
    * vscode에서는 코드를 작성할 수 있는 편집기, 파일을 탐색할 수 있는 탐색기 등 다양한 기능을 사용할 수 있으며, 이런 통합 개발 환경을 영어로 IDE(Integrated Development Environment)라고 합니다.
1. vscode 설치 및 설정하기
    1. [vscode 공식 홈페이지에 접속한다.](https://code.visualstudio.com/)
    2. 운영체제에 맞춰 파일을 다운로드한다.
    3. 라이선스 동의 후 계속 '다음' 버튼을 눌러 프로그램을 설치한다. (옵션을 따로 선택하지 않으셔도 됩니다)

    5. 한국어 설치가 완료되면 다시 vscode를 실행해 한국어를 적용한다.

## 웹 구조의 이해
> 웹에서 우리가 원하는 데이터를 마음껏 가져오기 위해선 기본적인 웹에 대한 이해가 필요합니다.

* 네이버, 구글과 같은 웹 페이지는 크게 HTML, CSS, JS라는 세 가지로 구성되어 있습니다.
    * 집으로 예를 들자면, 집의 뼈대를 만드는 것이 HTML입니다.
    * 뼈대와 방 구역에 대한 정의가 완성된 이후, 인테리어를 하는 것이 CSS입니다.
    * 인테리어 이후 사람이 들어와서 창문도 열거나 닫고, 조명을 켜거나 끄고, 태극기를 올리거나 내리는 등 집을 생동감 있게 만드는 것이 JS입니다.
    * 웹 스크래핑을 할 때는 HTML 정도만 간단히 알아도 괜찮습니다.
