# ③ 웹 스크래핑을 소개합니다.
> 이번에는 웹 스크래핑이 무엇인지, 어떤 것을 만들 수 있는지 배워봅시다.

* 웹 스크래핑(Web scraping)은 인터넷에 있는 웹 사이트에서 내가 원하는 정보를 자동으로 모아주거나, 저장해 주는 프로그램을 말합니다. 
    * Web은 World Wide Web의 줄임말이며, URL 가장 앞에 붙는 www가 World Wide Web에서 유래했습니다.
    * 인터넷은 컴퓨터 네트워크 통신망을 의미합니다.
    * 웹은 인터넷상에서 동작하는 서비스를 의미합니다.
* 웹 스크래핑 프로그램을 만들면 데이터를 수집하고, 분석하는 데 매우 유용합니다.
    * 예를 들어 상품 가격을 비교하는 웹 사이트에서 상품 가격 정보를 스크래핑하여 비교 정보를 제공하거나, 실시간으로 SNS의 댓글을 수집할 수 있습니다.

## 웹 스크래핑 VS 웹 크롤링
* 웹 스크래핑과 비슷한 개념으로 웹 크롤링(Web crawling)이 있습니다.
    * 웹 스크래핑은 인터넷상에서 내가 필요한 특정 데이터를 수집하고 분석하는 것에 초점을 맞춘 행위를 말합니다.
    * 웹 크롤링은 인터넷상에 있는 모든 웹 페이지를 탐색하고 필요한 정보를 수집하는 것에 초점을 맞춘 행위를 말합니다.
    * 웹 크롤링은 웹 크롤러(Web Crawler)라는 일종의 봇(bot)을 이용하여 웹 페이지 안에 있는 모든 페이지를 탐색하며 정보를 수집합니다. (선탐색 후정보추출)
    * 반면, 웹 스크래핑은 내가 데이터를 추출하고 싶은 특정 사이트가 있고, 그 사이트에서 특정 데이터를 수집합니다. (선정보결정 후탐색)
* 엄밀히 따지자면 둘은 분리되어야 하는 개념이지만, 편의상 둘을 구분하지는 않겠습니다.
    * 두 행위에 사용되는 Python 스킬도 차이가 거의 없습니다.
<br>

## 웹 스크래핑과 윤리
* 웹 스크래핑은 특정 웹 사이트에 접속하고, 내가 원하는 데이터를 내려받는 행위가 전부입니다. 따라서 기본적으로 불법은 아닙니다.
    * 하지만 불법으로 간주되는 경우가 정말 많으니, 이 기술은 매우 조심해서 사용해야 합니다. (특히 회사)
    * 뉴스 기사 : [여기어때가 야놀자 숙박정보 대량 빼갔다...2심도 ‘크롤링’ 불법 판정](http://www.newsroad.co.kr/news/articleView.html?idxno=18503)
    * 해당 사례를 요약하자면, 여기어때가 야놀자 데이터를 무단으로 수집해서 자기 회사의 데이터로 복제 후 이용하고 불법으로 판결이 났습니다.

<br>

<img src="https://user-images.githubusercontent.com/36785390/235336689-1f635aa6-1e06-4a0e-bccb-65d5540d0d15.jpg" width="28%">
  
이미지 출처 : http://prowebscraping.com/web-scraping-vs-web-crawling/


* 웹 스크래핑은 수집한 데이터를 상업적으로 이용하는 경우 혹은, 내 프로그램이 상대 서버에 부하를 주는 경우 불법 행위로 간주됩니다.
    * 상업적 목적이 아니라고 하여도 합법이라는 뜻은 아닙니다.

<br>

* 기본적으로 모든 웹 사이트들은 이런 웹 스크래핑 행위에 대한 규제 여부를 명시해 놓습니다.
    * 사이트별로 스크래핑, 크롤링 행위에 대한 여부는 robots.txt라는 파일에 작성해 둡니다.
    * 이와 관련한 내용은 다음에 웹 스크래핑을 실습하며 확인해 보겠습니다.

## HTML의 기본 용어
* HTML은 '<'로 시작해서 '>'로 끝나는 **태그(Tag)** 라는 것을 기준으로 동작합니다.
* 태그는 시작 태그와 종료 태그로 구성되어 있습니다.
    * 시작 태그의 예시 : \<html>
    * 종료 태그의 예시 : \</html> ('/' 기호가 앞에 붙음)
* 태그 사이에 다른 태그 및 텍스트를 작성하여 홈페이지를 만드는 것이 HTML입니다.

<img src="https://user-images.githubusercontent.com/36785390/235337150-3aaae2e5-c67c-41c2-bfd3-44b12943b88a.png" width="50%">

* 한 쌍의 태그는 **요소(Element)** 라고 부릅니다. 
* 요소는 시작 태그와 종료 태그로 이루어진 모든 HTML 명령어를 의미합니다.

<br>

* study.html 파일에 다음과 같이 코딩해봅시다.

``` html
    </head>
```

    * head : html 문서에 관한 기본 정보를 포함하고 있는 부분. 문서의 제목이나 제작자 등을 쓸 수 있음.
    * body : 실제 웹 페이지로 드러날 html의 본문을 작성하는 부분. html 코드의 대부분을 차지함.

    * h1 : html 본문에서 문단의 제목 역할을 할 수 있는 굵고 큰 글씨의 태그를 의미함.

* 또한, title과 h1의 시작 태그 - 종료 태그 사이에 작성된 글은 **콘텐츠(Content)** 라고 부릅니다.

``` python
t4 = s1 * n1
t5 = s2 * n2
# t6 = s1 + n1 이지만, 에러가 발생하므로 t6은 만들지 않는다.

print(t1, type(t1))
print(t2, type(t2))
print(t3, type(t3))
print(t4, type(t4))
print(t5, type(t5))
```

* type()은 '소괄호()'안에 변수를 넣어주면 그 변수가 어떤 자료형인지 알려주는 기능이다.
* 이 코드를 실행해보면 다음과 같은 사실을 알 수 있다.
    1. 변수의 연산 결과를 변수에 넣을 수 있다.
    2. 문자열 형태의 숫자는 더하기 혹은 곱하기를 했을 때 '숫자'가 더해지지 않고, 글자가 확장된다.
    3. 문자열에 숫자 변수를 곱하면 문자열이 만들어진다.
    4. 문자열 변수와 숫자 변수는 더할 수 없다.

* tip. 문자열 자료형은 str, 숫자형 자료형은 int라고 한다. (숫자 중에서도 정수만 int로 분류함. 나중에 더 자세히 배울 예정!)

### (더보기) 변수 생성의 규칙과 연산자 우선순위표
* 변수의 이름은 마음대로 만들 수 있지만, 아래의 규칙은 지켜주셔야 합니다.
    1. 영문 문자와 숫자를 사용합니다.
    2. 대소문자를 구분합니다. (변수 a와 A는 다름)
    3. 숫자부터 시작하면 안 됩니다. (문자부터 시작해야 함, 1a는 불가, a1은 가능)
    4. 특수 문자는 사용할 수 없으나, 밑줄 문자는 사용할 수 있습니다. (a_1, _a1 가능)
    5. 몇 가지 문자(키워드)는 사용할 수 없습니다. (print, input, str, if, sum, ...)

* 연산자 우선순위 표

| 우선순위 | 연산자 | 설명 |
|---|---|---|
| 1 | `**` | 지수 연산 |
| 2 | `*, /, %` | 곱셈, 나눗셈, 나머지 연산 |
| 3 | `+, -` | 덧셈, 뺄셈 연산 |

* 위의 표는 파이썬에서 연산자들의 우선순위를 보여줍니다. 이 외에도 다양한 연산자가 있지만, 더 배우면 오히려 혼동스러우므로, 지금은 배우지 않도록 하겠습니다!

## 심화 입출력 (함수란?)
> print와 input의 활용법을 조금 더 자세하게 배워봅시다.

* 화면에 데이터를 출력할 때 사용하는 코드는 print()였습니다.
    * 앞으로는 **영어 이름 + ()** 구조로 되어있는 것을 **함수(Function)** 라고 부르겠습니다. 
        * ex. print() -> print 함수

* print 함수는 쉼표 기호 ','를 이용해서 여러 개의 값을 출력할 수 있습니다.

````python
print(1, 2, 3)
````

* 실행 결과를 살펴보면 ','로 구분되는 데이터 사이에 자동으로 공백(띄어쓰기) 한 개가 추가되는 것을 알 수 있습니다!

* 이번에는 print 함수 세 개를 이용해서 똑같은 값을 화면에 출력해보겠습니다.

````python
print(1)
print(2)
print(3)
````

* 실행 결과를 자세히 살펴보면, print 함수가 소괄호 안에 있는 값을 출력하고 한 줄을 바꿔주는 것을 알 수 있습니다!!!

* print 함수는 다음과 같은 두 가지 기본적인 특징이 있습니다.
    1. ','로 구분되는 데이터 : 두 데이터 사이에 공백이 자동으로 추가된다.
    2. print 함수가 끝나면 : 자동으로 줄을 바꿔준다.

* 이 두 가지 기본적인 특징을 바꾸기 위해선 **함수**의 개념과 **매개변수, 인자** 라는 것을 알아야 합니다.

    <img src="https://user-images.githubusercontent.com/36785390/235345376-fed77e7f-8077-4565-8e4b-69f062c53dd5.png" width="30%">

* (※수학 주의※) 함수라는 상자는 입력된 값을 이용해서 상자 안에 있는 똑같은 일을 수행하고, 어떤 결과물을 반환한다는 특징이 있습니다.
    * 예를 들어 위의 그림에 있는 함수는 입력된 값을 5배 증가시켜서 반환해준다는 특징이 있습니다.

* 위의 과정을 코딩으로 구현하기 위해선
    1. 값의 입력 : 입력으로 들어오는 값인 2는 소괄호'()'에 넣어준다.
    2. 함수의 이름 : 함수의 이름은 '입력 값 * 5'이다.
    3. 함수의 역할 : 이 함수는 소괄호에 있는 값(입력된 값)에 5를 곱한다.
    4. 결과물의 반환 : 곱한 값을 return 이라는 키워드를 이용해서 반환하며, 이때 10이라는 값이 반환된다.

* 이번에는 print 함수를 살펴보겠습니다.
* print 함수는 다음과 같이 구성되어 있습니다.
    1. 값의 입력 : 소괄호 안에 출력하고 싶은 값을 넣어준다. 이때, 여러 개의 값을 출력하고 싶으면 ','로 분리한다.
    2. 함수의 이름 : print
    3. 함수의 역할 : 이 함수는 입력된 값을 화면에 출력한다. 쉼표로 구분된 값 사이에는 공백 한 칸을 넣어주고, 출력이 끝나면 자동으로 한 줄을 바꿔준다.
    4. 결과물의 반환 : 반환되는 값은 없다. (화면에 출력되는 값과 반환되는 값은 다른 값입니다!)
    5. 추가 기능(옵션) : 값을 입력하는 곳에서 기본 옵션을 변경할 수 있음
        1. sep : 쉼표로 구분된 값 사이에 넣어줄 문자를 지정해준다. 기본은 공백 한 칸.
        2. end : 출력이 끝나면 자동으로 출력할 문자를 지정해준다. 기본은 한 줄 바꿈.
    
    <img src="https://user-images.githubusercontent.com/36785390/235345377-0c4cb60c-f73d-42fd-aac6-23483c9f7c72.png" width="45%">

* 위에서 사용된 sep과 end는 사용하는 꼴이 마치 변수에 값을 넣어주는 것과 같습니다. 다만 일반 변수와는 그 역할이 다르므로, 함수의 입력 부분에서 사용하는 변수는 **매개변수(Parameter)** 라고 부릅니다.
    * 그리고 매개변수에 넣어주는 실제 값을 **인자(Argument)** 라고 부릅니다.

````python
print("Hi", "Python", sep="_")
````

````python
print(1, end='') # ' 사이에 아무 문자도 넣지 않음
print(2, end='')
print(3, end='')
````

* 마지막으로 input() 함수는 다음과 같이 구성되어 있습니다.
    1. 값의 입력 : 문자열을 넣어주면 화면에 똑같은 문자열을 print 해준다. (어떤 값을 입력하라고 사용자에게 힌트를 주는 역할)
    2. 함수의 이름 : input
    3. 함수의 역할 : 입력된 값을 화면에 출력해준 뒤, 사용자가 값을 입력하면 그 값을 반환(return) 함
    4. 결과물의 반환 : 입력된 값을 반환함

````python
name = input("이름을 입력해주세요 :")
age = input("나이를 입력해주세요 :")
print("저의 이름은 ", name, "입니다.", sep='')
print("저의 나이는", age, "살 입니다", end='.')
````
